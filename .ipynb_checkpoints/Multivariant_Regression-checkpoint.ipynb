## Assumptions
The code below is to verify for regression assumptions and fit a line to the data. These assumptions include:
- No multicollinearity
- Homoscedasticity
- No auto-correlation
- Normal distribution of error

```python
final_df.describe()
```

```python
# Remove outliers
final_df_remove_outliers = final_df[(np.abs(zscore(final_df)) < 3).all(axis=1)]

final_df.index.difference(final_df_remove_outliers.index)
```

```python
# Normalize columns as to not deal with different units
final_df_remove_outliers = final_df_remove_outliers / final_df_remove_outliers.max()
```

### Multicollinearity

```python
# Give us a better idea of multicollinearity
plt.figure(figsize=(20,17))
sns.heatmap(final_df_remove_outliers.corr(),vmax=1, annot=True,cmap='viridis')
plt.show()
```

```python
# Use a VIF table to get better idea of multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
def create_vif(dataframe):
    ''' This function calculates the Variation Inflation Factors for each column and convert it into a dataframe'''
    
    #create an empty dataframe
    vif_table = pd.DataFrame()
    #populate the first column with the columns of the dataset
    vif_table['variables'] = dataframe.columns
    #calculate the VIF of each column and create a VIF column to store the number
    vif_table['VIF'] = [vif(dataframe.values, i) for i in range(dataframe.shape[1])]
    
    return vif_table

#print the VIF table for each variable
print(create_vif(final_df_remove_outliers))
```

As for the assumption of multicollinearity we see that our Temperature column is higher for the VIF. This strikes use as concerning but also makes sense because the other factors are influenced by temperature. So we will keep this in the regression.

```python
# Plotting a pairplot
pairplot = sns.pairplot(final_df_remove_outliers, diag_kind="hist",  corner=True)
pairplot.map_lower(sns.kdeplot, levels=4, color=".2")
```

From the scatter plots above we can see the types of relationships between our features and predictor. From this we mostly see little to no relationship meaning that fitting a regression line will be more of a challenge. We do see that the Rain produces an inverse logistic curve reflected over the y axis. This variable will have to be kept in mind.

### Linear Regression Model

```python
# Perform the Linear Regression
feature_cols = ['Temp', 'Humidity', 'Wind', 'Rain']
X = final_df_remove_outliers[feature_cols]
Y = final_df_remove_outliers.ISI

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20, random_state=0)


# instantiate and fit
lm2 = LinearRegression()
lm2.fit(X_train, Y_train)

# print the coefficients
print('Y-Intercept:',lm2.intercept_)
print('Feature Coefficients:',lm2.coef_)
```

From the normalized data we can see the y-intercept being somewhat high for normalized data but have to know this doesnt necessarily spark any concern.

```python
# Display coefficients for each column
list(zip(feature_cols, lm2.coef_))
```

```python
# Predict the first 5 ISI values
y_predict = lm2.predict(X_test)

y_predict[:5]
```

```python
X2 = sm.add_constant(X)

lm1 = sm.OLS(Y,X2)

est = lm1.fit()
# Print the coefficients
est.params
```

```python
est.summary()
```

From the OLS Regression results we see that we get a poor R-Squared value which was expected since we dont really have a strong linear relationship from the independent and dependent variables. The testing was to experiement with the data and not necessarily have perfect results. To see if these results are trustworthy we will take a look at the other assumptions below.

```python
# Seeing some of the residuals and relationship between ISI and Temp
figsize = plt.figure(figsize=(20, 10))
fig = sm.graphics.plot_regress_exog(est,
                                    'Temp',
                                    fig=figsize)
```

### Homoscedasticity
The test for Homoscedasticity failed to accept the null hypothesis that the variances for the residuals are equal scattered meaning we have heteroscedasticity.

```python
_,pval,_,fpval = sm.stats.diagnostic.het_white(est.resid, est.model.exog)
print(pval,fpval)
```

### Auto Correlation
The test for auto correlation also failed to accept the null hypothesis meaning we do not have auto correlation

```python
lag = min(10,(len(X)//5))
test_results = sm.stats.diagnostic.acorr_ljungbox(est.resid, lags= lag, return_df=True)
test_results
# Failed to reject null hypothesis so there is no auto correlation
```

### Normal Distribution in Residuals
The test for the residuals to be normally distributed also seems to be true since we can see them visually being close to the line in the qqplot.

```python
sm.qqplot(est.resid, line= 's')
mean_residuals = sum(est.resid) / len(est.resid)
mean_residuals
```

## Accuracy Measurements
Our assumptions were tested and we can say that we can trust the linear regression model that we have created

### Accuracy Values

```python
# Getting the MSE MAE and RMSE
model_mse = metrics.mean_squared_error(Y_test,y_predict)

model_mae = metrics.mean_absolute_error(Y_test,y_predict)

model_rmse = math.sqrt(model_mse)

print("MSE {:.3}".format(model_mse),"MAE {:.3}".format(model_mae),"RMSE {:.3}".format(model_rmse))
```

```python
model_r2 = metrics.r2_score(Y_test, y_predict)
model_r2
```

From the values we can see that we have smaller MSE and MAE values which is good but we see a lower R-squared value. This tells us that our model although correct doesnt do the best at predicting the ISI based off the independent variables

```python
# Getting an understanding of which features should be used to use in the linear regression using cross validation
model = LinearRegression(fit_intercept=False)
selector = RFECV(model,cv=5)
selector = selector.fit(X, Y)
print(f"Out of {len(X.columns)} features, best number of features {selector.n_features_}")

plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score")
plt.plot(range(1, len(X.columns) + 1), selector.grid_scores_)
print(X.columns[selector.support_].values)
plt.show()
```

```python
# Applying a similar test as before but this time using the R-Squared score
mask = selector.support_
print(f"Best features according to RFE {X.columns[mask].values}")

X_m1 = X.iloc[:,mask]
model1 = LinearRegression().fit(X_m1,Y)
print(f"R2 Score: {model1.score(X_m1,Y)}")
```

From the cross-validation plot and RFE we see the best score is to use all 4 features. One feature we were concerned about was Rain but looks like it is still best to include it

```python
# Plotting Forward Selection to see which features are best to keep
model = LinearRegression(fit_intercept=False)
sfs1 = sfs(model,k_features=4,forward=True,scoring='r2',cv=5)
sfs1.fit(X,Y)
fig = plot_sfs(sfs1.get_metric_dict())
plt.title('Forward Selection')
plt.grid()
plt.show()
```

```python
index = list(sfs1.k_feature_idx_)
X_m1 = X.iloc[:,index]
model1 = LinearRegression().fit(X_m1,Y)
print(f"R2 Score: {model1.score(X_m1,Y)}")
```
