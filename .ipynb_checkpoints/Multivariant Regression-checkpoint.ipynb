# Multivariant Regression

```python
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from sklearn.linear_model import Lasso
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore
from sklearn.model_selection import train_test_split
```

## Assumptions
The code below is to verify for regression assumptions and fit a line to the data. These assumptions include:
- No multicollinearity
- Homoscedasticity
- No auto-correlation
- Normal distribution of error

```python
final_df.describe()
```

```python
plt.figure(figsize=(20,17))
sns.heatmap(final_df.corr(),vmax=1, annot=True,cmap='viridis')
plt.show()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif
def create_vif(dataframe):
    ''' This function calculates the Variation Inflation Factors for each column and convert it into a dataframe'''
    
    #create an empty dataframe
    vif_table = pd.DataFrame()
    #populate the first column with the columns of the dataset
    vif_table['variables'] = dataframe.columns
    #calculate the VIF of each column and create a VIF column to store the number
    vif_table['VIF'] = [vif(dataframe.values, i) for i in range(dataframe.shape[1])]
    
    return vif_table

#print the VIF table for each variable
print(create_vif(final_df))
```

```python
pairplot = sns.pairplot(final_df, diag_kind="hist",  corner=True)
pairplot.map_lower(sns.kdeplot, levels=4, color=".2")
```

```python
final_df_remove_outliers = final_df[(np.abs(zscore(final_df)) < 3).all(axis=1)]

final_df.index.difference(final_df_remove_outliers.index)
```

```python
pairplot = sns.pairplot(final_df_remove_outliers, diag_kind="hist",  corner=True)
pairplot.map_lower(sns.kdeplot, levels=4, color=".2")
```

From the scatter plot above we can see the types of relationships between our features and predictor. From this we mostly see little to no relationship meaning that fitting a regression line will be more of a challenge. We do see that the FFMC produces an inverse logistic curve while the Rain feature produces the same curve but reflected over the y axis. These two variables will have to be kept in mind. We also see one clear outlier that will need to be dealt with.

After fitting a line to the data we get an intercept of 10. This tells us that when every other feature is zero we will get a positive ISI value which makes sense. For example our temperature would be zero which wouldnt help any fire spread. As for the coefficients for our features they are all positive except for humidity. From our scatterplot we see a slightly negative relationship.

```python
feature_cols = ['Temp', 'Humidity', 'Wind', 'Rain']
X = final_df_remove_outliers[feature_cols]
Y = final_df_remove_outliers.ISI

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.20, random_state=0)


# instantiate and fit
lm2 = Lasso()
lm2.fit(X_train, Y_train)

# print the coefficients
print(lm2.intercept_)
print(lm2.coef_)
```

```python
# Display results for each column
list(zip(feature_cols, lm2.coef_))
```

```python
# Predict values
y_predict = lm2.predict(X_test)

y_predict[:5]
```

```python
X2 = sm.add_constant(X)

lm1 = sm.OLS(Y,X2)

est = lm1.fit()
# Print the coefficients
est.params
```

```python
est.summary()
```

```python
figsize = plt.figure(figsize=(20, 10))
fig = sm.graphics.plot_regress_exog(est,
                                    'Temp',
                                    fig=figsize)
```

```python
_,pval,_,fpval = sm.stats.diagnostic.het_white(est.resid, est.model.exog)
_,pval_pag,_,fpval_pag = sm.stats.diagnostic.het_breuschpagan(est.resid, est.model.exog)
print(pval,fpval)
print(pval_pag,fpval_pag)
# fail to reject that the variances for the errors are equal...Meaning we have no heteroscedasticity
```

```python
lag = min(10,(len(X)//5))
test_results = sm.stats.diagnostic.acorr_ljungbox(est.resid, lags= lag, return_df=True)
test_results
# Failed to reject null hypothesis so there is no auto correlation
```

```python
sm.qqplot(est.resid, line= 's')
mean_residuals = sum(est.resid) / len(est.resid)
mean_residuals
```

Everything in model is working nicely

```python
import math

model_mse = metrics.mean_squared_error(Y_test,y_predict)

model_mae = metrics.mean_absolute_error(Y_test,y_predict)

model_rmse = math.sqrt(model_mse)

print("MSE {:.3}".format(model_mse),"MAE {:.3}".format(model_mae),"RMSE {:.3}".format(model_rmse))
```

```python
model_r2 = metrics.r2_score(Y_test, y_predict)
model_r2
```
